## All commands at a glance:

1. git init
Initialize a new Git repository.

2.git clone <repository_url>
Clone a remote repository to your local machine.

3. git status
Check the current status of your repository.

4.git add <file>
Stage changes in a file for the next commit.

5.git add .
Stage all changes in the current directory.

6.git commit -m "Your commit message"
Commit staged changes to the repository with a message.

7.git pull
Fetch and merge the latest changes from a remote repository.

8.git push
Push local commits to a remote repository.

9.git branch <branch_name>
Create a new branch.

10.git checkout -b <branch_name>
Create and switch to a new branch.

11.git checkout <branch_name>
Switch to an existing branch.

12.git merge <branch_name>
Merge changes from one branch into the current branch.

13.git log
View a log of all commits in the current branch.

14.git diff
Show differences between the working directory and the last commit.

15.git checkout -- <file>
Discard changes in a file.

16.git reset --hard
Discard all changes in the working directory.

17.git clean -n
Show untracked files that will be removed.

18.git clean -f
Remove untracked files.

19.git reset HEAD~1
Undo the last commit while keeping changes in the working directory.

20.git stash
Temporarily save changes that are not ready to be committed.

21.git stash apply
Apply the most recent stash.

22.git stash pop
Apply and remove the most recent stash.
