#### AI-Enabled software development methodology introduction [Agile, Squad, waterfall, etc]

1. Agile:

Principles:

- Iterative and Incremental: Agile development is characterized by short development cycles, called iterations or sprints, where small portions of a project are completed in each cycle.
- Collaboration: Emphasis is placed on collaboration and communication between cross-functional teams, including developers, testers, and business stakeholders.
- Customer Feedback: Agile incorporates frequent feedback from customers or end-users throughout the development process, allowing for adjustments and improvements.
- Adaptability: Agile is known for its adaptability to changing requirements, even late in the development process.


Advantages:

- Faster time-to-market due to incremental development.
- Enhanced adaptability to changing requirements.
- Continuous customer involvement fosters better product alignment with user needs.


Disadvantages:

- May require a cultural shift within organizations.
- Limited predictability in terms of project timelines and costs.


Squad (Spotify Model):

Overview:

- The Squad model is a concept popularized by Spotify to organize development teams.
- Cross-functional teams, referred to as squads, work autonomously on specific features or components.
- Each squad is self-contained, with members possessing diverse skills needed to handle end-to-end development for their assigned area.


Characteristics:

- Autonomy: Squads have the autonomy to make decisions related to their assigned features or components.
- Cross-functionality: Each squad comprises members with different skills, such as developers, testers, designers, and product owners.
- Alignment: Squads align with the overall product vision and strategy.


Advantages:

- Faster decision-making within autonomous teams.
- Efficient handling of specific feature sets.

Disadvantages:

- May require strong communication and coordination to avoid silos.

Waterfall:

Process Flow:

- Waterfall is a traditional, linear approach to software development.
- Development is divided into sequential phases: Requirements, Design, Implementation, Testing, Deployment, and Maintenance.
- Each phase must be completed before moving on to the next.

Advantages:

- Well-defined structure and documentation at each stage.
- Predictable and easy to manage.

Disadvantages:

- Limited flexibility for changes after the project has started.
- Potential for long delivery times before tangible results are seen.
- Customer feedback is typically incorporated late in the process.

Use Cases:

- Waterfall is often suitable for projects with well-defined and stable requirements.
- Commonly used in industries with strict regulatory compliance requirements.

## MLOps Motivation: High-level view

- Motivation: MLOps aims to streamline and automate the end-to-end machine learning lifecycle, from model development to deployment and monitoring. It addresses challenges related to collaboration, reproducibility, and scalability in ML projects.

### MLOps Challenges:


#### Scalability:

- MLOps helps ensure that ML models are able to scale efficiently and can handle large volumes of data and user requests when needed.
- This is utterly important for applications that require real-time decision-making or processing of high-velocity data streams.

#### Reliability
- With the help of MLOps, ML models should be by all means reliable and deliver consistent results over time. 
- This is important for applications that require high accuracy, first and foremost, fraud detection and predictive maintenance.

#### Security
- One of MLOps goals is to make ML models secure and protected against threats, including data breaches, cyber-attacks, and malicious actors that endanger applications dealing with sensitive or confidential data.
- IAM

#### Cost-efficiency
- MLOps helps to optimize the use of resources such as computing power, storage, and data bandwidth by automating processes and reducing manual labor.
- This can lead to significant cost savings for businesses that rely on ML models for decision-making and analytics.
